Making tables

```python
# Connecting to the MySQL server
conn = mysql.connector.connect(user='root', password='1734680nN', host='localhost', port=3306)

# Making a cursor object to interact with the database
cursor = conn.cursor()

# naming the database
db_name = "police_shootings"

# Creating the database
cursor.execute(f"CREATE DATABASE IF NOT EXISTS {db_name}")

cursor.execute(f"USE {db_name}")

# Creating tables
create_person_table = """
CREATE TABLE IF NOT EXISTS Person (
    id INT AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(255),
    age INT,
    gender VARCHAR(10),
    race VARCHAR(50)
)
"""

create_location_table = """
CREATE TABLE IF NOT EXISTS Location (
    id INT AUTO_INCREMENT PRIMARY KEY,
    city VARCHAR(100),
    state VARCHAR(2),
    longitude DECIMAL(9,6),
    latitude DECIMAL(8,6),
    is_geocoding_exact BOOLEAN
)
"""

create_incident_table = """
CREATE TABLE IF NOT EXISTS Incident (
    id INT AUTO_INCREMENT PRIMARY KEY,
    person_id INT,
    location_id INT,
    date DATE,
    manner_of_death VARCHAR(255),
    armed VARCHAR(255),
    signs_of_mental_illness BOOLEAN,
    threat_level VARCHAR(50),
    flee VARCHAR(50),
    body_camera BOOLEAN,
    FOREIGN KEY (person_id) REFERENCES Person(id),
    FOREIGN KEY (location_id) REFERENCES Location(id)
)
"""

# executing the creation of the tables
cursor.execute(create_person_table)
cursor.execute(create_location_table)
cursor.execute(create_incident_table)

# Committing the changes
conn.commit()

print("Tables created successfully!")
```

```python
# Populating the tables
# Reading in the clean data from CSV
df = pd.read_csv('US Police shootings in from 2015-22_Cleaned.csv')

# Putting the data into the table
for _, row in df[['name', 'age', 'gender', 'race']].iterrows():
    cursor.execute("""
    INSERT INTO Person (name, age, gender, race)
    VALUES (%s, %s, %s, %s)
    """, (row['name'], row['age'], row['gender'], row['race']))

# Putting the data into the table
for _, row in df[['city', 'state', 'longitude', 'latitude', 'is_geocoding_exact']].iterrows():
    cursor.execute("""
    INSERT INTO Location (city, state, longitude, latitude, is_geocoding_exact)
    VALUES (%s, %s, %s, %s, %s)
    """, (row['city'], row['state'], row['longitude'], row['latitude'], row['is_geocoding_exact']))

# Committing the changes
conn.commit()

# Putting the data into the table
for _, row in df[['date', 'manner_of_death', 'armed', 'signs_of_mental_illness', 
                  'threat_level', 'flee', 'body_camera']].iterrows():
    cursor.execute("""
    INSERT INTO Incident (date, manner_of_death, armed, signs_of_mental_illness, 
                          threat_level, flee, body_camera)
    VALUES (%s, %s, %s, %s, %s, %s, %s)
    """, (row['date'], row['manner_of_death'], row['armed'], row['signs_of_mental_illness'],
          row['threat_level'], row['flee'], row['body_camera']))

# Committing the changes
conn.commit()

print("Tables populated successfully!")
```

```python
# Querying to see if the tables were populated 
# Query for some sample data from the person table
cursor.execute("SELECT * FROM Person LIMIT 5")
person_data = cursor.fetchall()
print("Sample data from 'Person' table:")
for row in person_data:
    print(row)
# Query for some sample data from the laction table
cursor.execute("SELECT * FROM Location LIMIT 5")
location_data = cursor.fetchall()
print("\nSample data from 'Location' table:")
for row in location_data:
    print(row)

# Query for some sample data from the incident table
cursor.execute("SELECT * FROM Incident LIMIT 5")
incident_data = cursor.fetchall()
print("\nSample data from 'Incident' table:")
for row in incident_data:
    print(row)
```
